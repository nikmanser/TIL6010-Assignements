### 8. Is there any common dates (or periods) among these peaks (and/or valleys) of different activities

*Food for thoughts*:
- One of the most naive way to find the overlapping dates is to look for common indexes (or dates) of the peaks/valleys
- A more general solution is to have a tolerance level between the days. For example, Sep 19 and Sep 20 can be considered matched (or overlapped) if the tolerance level is at least one day.

```python
#Let's find the common days of peak for Residential and Workplaces
#We'll deal with the tuples ilocs_max_res and ilocs_max_wp

res_set = set(ilocs_max_res)     #we put in a set to isolate and then compare every element in the tuple, for both the tuples
wp_set = set(ilocs_max_wp)

common_peaks = (res_set and wp_set)  #we find the common records (which are referred to a single day) with a peak in that week

dates = []                          #with a for loop and loc function we can find the date which the records mentioned are referred to
for i in common_peaks:
    date = df_2021.loc[i, 'date']
    dates.append(date)
    

print(dates)
```
